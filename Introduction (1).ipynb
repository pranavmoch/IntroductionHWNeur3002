### Integrate and Fire Model: An IF model contains and integrator and a spike generator. As current is injected into the system, the integrator sums the inputs and the spike generator creates the spike with the certain voltage, leading to the output. This is where is is determined if the voltage surpasses the threshold or not to create and action potential or not.

##### Below is our code for the IF Model

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
time_step = 0.1  # ms
total_time = 200  # ms
time = np.arange(0, total_time, time_step)

# Membrane parameters
resting_potential = -70  # mV
threshold_potential = -55  # mV
reset_potential = -75  # mV
input_current = 10  # pA
membrane_resistance = 10  # M
membrane_time_constant = 20  # ms

# Initialize variables
membrane_potential = np.ones(len(time)) * resting_potential
spike_train = np.zeros(len(time))

# Simulation loop
for t in range(1, len(time)):
    if membrane_potential[t - 1] >= threshold_potential:
        # Spike and reset potential
        membrane_potential[t - 1] = 30  # Spike (not biologically precise, illustrative only)
        membrane_potential[t] = reset_potential
    else:
        # Update membrane potential
        dV = (
            -(membrane_potential[t - 1] - resting_potential) + membrane_resistance * input_current
        ) * (time_step / membrane_time_constant)
        membrane_potential[t] = membrane_potential[t - 1] + dV

# Plot results
plt.figure(figsize=(10, 6))
plt.plot(time, membrane_potential, label="Membrane Potential (mV)", color="b")
plt.axhline(threshold_potential, color="r", linestyle="--", label="Threshold (-55 mV)")
plt.axhline(resting_potential, color="g", linestyle="--", label="Resting Potential (-70 mV)")
plt.title("Integrate-and-Fire Neuron Model")
plt.xlabel("Time (ms)")
plt.ylabel("Membrane Potential (mV)")
plt.legend()
plt.grid()
plt.show()
```

#### The above visualization shows the IF model. The limitation of this is that it does not account for the leaky ion channels. The LIF model on the other hand does. The limitation that the LIF model has is that it does not include voltage gated channels. In contrast, the LIF model is similar to the biological neuron because it incorporates leaky ion channels allowing for a natural gradual decay of voltage back to resting. The LIF model is also similar to the neuron becuase it integrates inputs and then it uses the spike generator to determine if an action potential fires or not.
